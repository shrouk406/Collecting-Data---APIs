```python
currencies = []
rates = []
```

```python
for currency, rate in exchange_rates.items():
    print(f"{currency} {rate}")
    currencies.append(currency)
    rates.append(rate)
    
```

```python
with open("currencirs exchange.csv", mode='w', encoding='utf-8') as fd:
    fd.write("id,currency,rate with EUR\n") # header
    # body
    for i  in range(len(currencies)):
        e_cuurency = currencies[i]
        e_rate = exchange_rates[e_cuurency]
        fd.write(f"{i+1},{e_cuurency},{e_rate}\n")
```
